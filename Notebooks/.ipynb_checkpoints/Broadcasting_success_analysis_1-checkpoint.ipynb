```python
import pandas as pd

excel_file = pd.ExcelFile('../Data/Broadcast_Tracking_ARCHIVE.xlsx')

# Print sheet names and types
for sheet_name in excel_file.sheet_names:
    print(f"Sheet Name: {sheet_name}, Type: {type(sheet_name)}")

# Attempt to parse sheets into a dictionary of DataFrames
dfs = {}
for sheet_name in excel_file.sheet_names:
    try:
        dfs[sheet_name] = excel_file.parse(sheet_name)
    except ValueError as e:
        print(f"Skipping sheet '{sheet_name}' due to error: {e}")
```

```python
# Define a mapping of old sheet names to new sheet names
sheet_name_mapping = {
    'Monthly Maxes': 'Monthly_Maxes',
    'Re-work Broadcast Tracking': 'Broadcast_Tracking',
    'Nigeria - NTA': 'Nigeria_NTA',
    'RTV-Rwanda': 'RTV_Rwanda',
    '2014-15': '2014_15',
    'Quarterly Reach': 'Quarterly_Reach',
    'Reach Calculations - Aug. 2017': 'Reach_Calculations',
    'Broadcast Schedule': 'Broadcast_Schedule',
    'Ubongo Kids - TBC1': 'Ubongo_Kids',
    'Kilimani Sesame': 'Kilimani_Sesame',
    'Akili and Me - TBC1': 'Akili_and_Me',
    'CITIZEN TV - Ubongo Kids': 'CITIZEN_TV_Ubongo_Kids',
    'CITIZEN TV- Akili': 'CITIZEN_TV_Akili',
    'Zambia - ZNBC': 'Zambia_ZNBC',
    'NTV Uganda': 'NTV_Uganda',
    'equip t stories': 'equip_stories',
    'Akili Radio': 'Akili_Radio',
    'NTV Kenya': 'NTV_Kenya',
    'NTV - ARCHIVE': 'NTV_ARCHIVE',
    'YouTube - Webisode': 'YouTube_Webisode',
    'Umuhimu wa Elimu': 'Umuhimu_wa_Elimu'
}

# Create an empty dictionary to store the renamed DataFrames
renamed_dfs = {}

for sheet_name, original_df in dfs.items():
    # Use the mapping to get the new sheet name or keep the original if not found
    new_sheet_name = sheet_name_mapping.get(sheet_name, sheet_name)

    # Add to the dictionary with the original sheet name
    renamed_dfs[new_sheet_name] = original_df
    
    if new_sheet_name in renamed_dfs:
        renamed_df = renamed_dfs[new_sheet_name]
```

```python
def read_pdf(file_path):
    with open(file_path, 'rb') as file:
        pdf_reader = PyPDF2.PdfReader(file)
        
        text = ''
        for page_number in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_number]
            text += page.extract_text()

    return text

# Replace 'your_pdf_file.pdf' with the actual PDF file path
pdf_file_path = '../Data/Akili and Me – Akr User Testing Findings - in progress .pdf'
pdf_text = read_pdf(pdf_file_path)

print(pdf_text)

```

#### Effectiveness of the programs to achieve its objective 

```python

# Pattern to match the desired paragraph
pattern = re.compile(r"The primary objective of the reading segment is to help children develop print awareness.*?Five Chicks and a Mango", re.DOTALL)

# Find the matched pattern
match = pattern.search(pdf_text)

# Extract the matched paragraph
if match:
    desired_paragraph = match.group(0)
    print(desired_paragraph)
else:
    print("Pattern not found.")
```
